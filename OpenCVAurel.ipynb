## Tracer une ligne

```python
img=cv.imread("C:/Users/pypar/Documents/M1 MMAA/Semestre 8/ProjetDeRecherche/Image/Foot.png",1)

img=cv.line(img,(0,0),(0,80),(0,0,255),5)
img=cv.line(img,(0,0),(80,0),(0,0,255),5)
img=cv.arrowedLine(img,(0,0),(80,80),(0,0,255),5)

plt.imshow(img[:,:,::-1])
plt.show()
```

## Relier plusieurs points

```python
H,W=img.shape[:2]

pts=np.array([[0,H-1],[W-1,H-1],[W-1,0]])
img=cv.polylines(img,[pts],True,(0,255,0),5)
plt.imshow(img[:,:,::-1])
plt.show()
```

## Tracer un Rectangle

```python
img=cv.imread("C:/Users/pypar/Documents/M1 MMAA/Semestre 8/ProjetDeRecherche/Image/Foot.png",1)

img=cv.rectangle(img,(25,60),(135,170),(0,0,255),2)
plt.imshow(img[:,:,::-1])
plt.show()
```

## Ecrire du texte

```python
img=cv.imread("C:/Users/pypar/Documents/M1 MMAA/Semestre 8/ProjetDeRecherche/Image/Foot.png",1)

img=cv.putText(img,"Ballon",(30,30),cv.FONT_HERSHEY_COMPLEX,1,(255,255,255),2)
plt.imshow(img[:,:,::-1])
plt.show()
```

## Faire un cercle

```python
img=cv.imread("C:/Users/pypar/Documents/M1 MMAA/Semestre 8/ProjetDeRecherche/Image/Foot.png",1)

img=cv.circle(img,(80,115),60,(255,255,255),2)
plt.imshow(img[:,:,::-1])
plt.show()
```
